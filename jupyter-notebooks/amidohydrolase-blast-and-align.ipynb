# Alignment

```python
os.chdir('/home/ecutts/Amidohydrolases/tree-NCBI-700-1200-hits/')

treeseqs = '../p-carrageenovora-tree.fasta'
# add representatives of the p carrageenovora sulfatase tree to the alignment
dry_reps = pd.read_table('tree-representatives/dry-representatives.txt', header=None)[0].tolist()
wet_reps = pd.read_table('tree-representatives/wet-representatives.txt', header=None)[0].tolist()
flo_reps = pd.read_table('tree-representatives/floating-representatives.txt', header=None)[0].tolist()
euk_reps = pd.read_table('tree-representatives/euk-representatives.txt', header=None)[0].tolist()

problem_seqs = pd.read_table('seqs-to-remove.txt', header = None)[0].tolist()
print(problem_seqs)
```

```python
seqs = [] # will contain all the sequences to be aligned including hits and reps
compare_seqs = []
compare_names =[]
```

```python
reps = dry_reps + wet_reps + flo_reps + euk_reps

with open(treeseqs, 'r') as read:
    for record in SeqIO.parse(read, 'fasta'):
        if record.id.split(':')[0] in reps and record.seq not in compare_seqs:
            if record.description not in problem_seqs:
                seqs.append(record)
                compare_seqs.append(record.seq)
                compare_names.append(record.id)
```

```python
with open('hits.fasta', 'r') as hits:
    for record in SeqIO.parse(hits, 'fasta'):
        if record.seq not in compare_seqs and record.id not in compare_names:
            if record.description not in problem_seqs:
                compare_seqs.append(record.seq)
                compare_names.append(record.id)
                seqs.append(record)
```

```python
outfile = 'amidohydrolase-superfamily-hits-and-reps.fasta'
with open(outfile, 'w') as out:
    SeqIO.write(seqs, out, 'fasta')
```

```python
mafft = '/cm/shared/engaging/mafft/7.245-with-extensions/bin/mafft' # path to mafft

import subprocess
def align(file): # function that performs alignment and prints process info
    output = open(file.split('.fasta')[0] + '.aln', 'w')
    print(
        subprocess.run([mafft, '--reorder', '--auto', file], stdout=output)
    )
    output.close()

align(outfile)
```
