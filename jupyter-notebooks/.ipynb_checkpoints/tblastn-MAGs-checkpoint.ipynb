## tBLASTn for rProts

```python
for MAG in os.listdir():
    if MAG.endswith('fa'): # only check the MAG databases, not random other things in the directory
        shortname = MAG.split('SB_biofilm_')[1]
        p = run([tblastn, '-query', 'queries/bfragilis-30rprots.faa',\
                '-db', MAG, '-out', 'results/rprots_vs_' + shortname,\
                '-outfmt', '6'], stdout=PIPE)
        print(p.stdout) # need to work on the stdout thing.
```

```python
os.chdir('results/')
```

```python
dfs = {}
header = ['query acc.ver', 'subject acc.ver', '% identity', 'alignment length', 'mismatches',\
          'gap opens', 'q. start', 'q. end', 's. start', 's. end', 'evalue', 'bit score']
for outfile in os.listdir():
    if outfile.endswith('.out'):
        name = outfile.split('.')[0].split('vs_')[1]
        dfs[name] = pd.read_table(outfile, names=header)
```

```python
os.listdir()
```

```python
# filter everything with an e value higher than 0.05
e05_dfs = {}
for MAG in dfs.keys():
    df = dfs[MAG]
    filtered_df = df[df['evalue'] <= 0.05]
    e05_dfs[MAG] = filtered_df
```

```python
# get names of rprots for query accessions
prot_names = {}
with open("../queries/bfragilis-30rprots.faa", "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        prot_names[str(record.id)] = record.description.split('protein ')[1].split(' [')[0]
```

```python
# add column to filtered df for rprots
for MAG in e05_dfs: 
    df = e05_dfs[MAG]
    df['rprot'] = df.apply(lambda row: prot_names[row['query acc.ver']], axis=1)
```

```python
with open('../../rprots/rprots_MAG_11.faa', "w") as out:
    rprot_count = {}
    seqs = []
    for index, row in e05_dfs['MAG_11'].iterrows():
        start = int(row['s. start'])-1
        end = int(row['s. end'])-1
        seq = row['subject acc.ver']
        rprot = row['rprot']
        if rprot not in rprot_count.keys():
            rprot_count[rprot] = 1
        else:
            rprot_count[rprot] +=1
        with open("../SB_biofilm_MAG_11.fa", "r") as handle:
            for record in SeqIO.parse(handle, "fasta"):
                if record.id == seq:
                    nucseq = record.seq
                    if end < start:
                        start = len(nucseq) - start -1 
                        end = len(nucseq) - end -1
                        nucsub = nucseq.reverse_complement()[start:end]
                        prot = nucsub.translate()
                    else: 
                        nucsub = nucseq[start:end]
                        prot = nucsub.translate()

        rec = SeqRecord(prot, id=seq + '_trns',name=rprot,description='Ribosomal Protein ' + rprot + '[MAG 11]')
        seqs.append(rec)

    print(seqs)
    SeqIO.write(seqs, out, 'fasta')
```

```python
def write_translations(name, nucfasta, df):
    with open('../../rprots/rprots_' + name + '.faa', "w") as out:
        rprot_count = {}
        seqs = []
        for index, row in df.iterrows():
            start = int(row['s. start'])-1
            end = int(row['s. end'])-1
            seq = row['subject acc.ver']
            rprot = row['rprot']
            if rprot not in rprot_count.keys():
                rprot_count[rprot] = 1
            else:
                rprot_count[rprot] +=1
            with open(nucfasta, "r") as handle:
                for record in SeqIO.parse(handle, "fasta"):
                    if record.id == seq:
                        nucseq = record.seq
                        if end < start:
                            start = len(nucseq) - start -1 
                            end = len(nucseq) - end -1
                            nucsub = nucseq.reverse_complement()[start:end]
                            prot = nucsub.translate()
                        else: 
                            nucsub = nucseq[start:end]
                            prot = nucsub.translate()
            neat_name = name.replace('_', ' ')
            rec = SeqRecord(prot, id=seq + '_trns',name=rprot,description='Ribosomal Protein ' + rprot + \
                        ' [' + neat_name + ']')
            seqs.append(rec)

        print(seqs)
        SeqIO.write(seqs, out, 'fasta')
```

```python
print(e05_dfs.keys())
```

```python
for MAG in e05_dfs.keys():
    name = MAG
    df = e05_dfs[MAG]
    nucfasta = '../SB_biofilm_' + name + '.fa'
    write_translations(name, nucfasta, df)
```

```python
counts = {}
rprots = pd.read_csv('/home/ecutts/metadata/rprots.txt', header = None)[0].tolist()
for faa in os.listdir():
    if faa.endswith('.faa'):
        name = faa.split('rprots_')[1].split('.faa')[0]
        counts[name] = dict.fromkeys(rprots)
        for prot in counts[name]:
            counts[name][prot] = 0
        with open(faa, 'r') as readfile:
            for record in SeqIO.parse(readfile, 'fasta'):
                prot = record.description.split('Protein ')[1].split(' [')[0]
                print(prot)
                counts[name][prot] += 1
                print(counts[name])
```

```python
for MAG in counts:
    print(MAG + ' count:' + str(sum(counts[MAG].values())))
    for rprot in counts[MAG].keys():
        if counts[MAG][rprot] != 1:
            print(rprot + ': ' + str(counts[MAG][rprot]))
    print('---')

    
```
