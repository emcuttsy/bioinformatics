```python
import pandas as pd
import os
```

```python
wdir = '/home/ecutts/MAGs-blastdb/results/'
os.chdir(wdir)
```

```python
MAG_metadata = pd.read_table('/home/ecutts/metadata/mags-meta.xls')
MAG_metadata.head()
```

```python
qlen = 1038
```

```python
header = ['query acc.ver', 'subject acc.ver', '% identity',
          'alignment length', 'mismatches', 'gap opens', 
          'q. start', 'q. end', 's. start', 's. end', 'evalue',
          'bit score']

dfs = {}
for outfile in os.listdir():
    if outfile.endswith('.out'):
        name = outfile.split('.')[0].split('_')[2]
        MAG = MAG_metadata.loc[MAG_metadata['taxon_oid'] == int(name)]['Genome Name / Sample Name'].values[0]
        phylum = MAG_metadata.loc[MAG_metadata['taxon_oid'] == int(name)]['Phylum'].values[0]
        
        dfs[name] = pd.read_table(outfile, header=None, names=header)
        dfs[name]['coverage'] =  (dfs[name]['q. end'] - dfs[name]['q. start'] + 1) / qlen
        # 1e-10 E value
        dfs[name] = dfs[name][dfs[name]['evalue'] <= 1e-10]
        # filter by percent identity 
        dfs[name] = dfs[name][dfs[name]['% identity'] >= 30]
        # filter by coverage
        dfs[name] = dfs[name][dfs[name]['coverage'] >= 0.8]
        dfs[name]['assembly'] = name
        dfs[name]['MAG'] = MAG
        dfs[name]['phylum'] = phylum
```

```python
combined = pd.concat(dfs.values(), ignore_index=True)
combined
```

```python
# get summary by MAG:
for MAG in set(combined['MAG']):
    print(MAG)
    print(list(combined.loc[combined['MAG'] == MAG]['phylum'])[0])
    print('count: ' + str(len(combined.loc[combined['MAG'] == MAG])))
    print()
```

```python
# Get summary by phylum
for phylum in set(combined['phylum']):
    print(phylum)
    print('count: ' + str(len(combined.loc[combined['phylum'] == phylum])))
    print('MAGs:')
    for MAG in set(combined.loc[combined['phylum'] == phylum]['MAG']):
        print(MAG)
    print()
```
