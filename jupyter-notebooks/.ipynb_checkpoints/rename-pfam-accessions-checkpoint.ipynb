```python
import pandas as pd
import numpy as np
from Bio import SeqIO
from ete3 import NCBITaxa
import os
```

```python
meta = pd.read_table('/home/ecutts/metadata/pfam-amidohydrolase-metadata.txt')
meta = meta[meta['Protein names'] != 'Deleted.'] #delete deleted sequences
#meta = meta[meta['Organism ID'] != None]

def neaten_taxid(orgid):
    return str(orgid).split('.0')[0]

meta['Organism ID'] = meta['Organism ID'].apply(neaten_taxid) # taxids now in NCBI format
```

```python
meta.head()
```

```python
infile = '/home/ecutts/Amidohydrolases/PFAM-amidohydrolase.fasta'
outfile = '/home/ecutts/Amidohydrolases/PFAM-amidohydrolase-neatened.fasta'
ncbi = NCBITaxa()
with open(infile, "r") as read, open(outfile, 'w') as write:
    for record in SeqIO.parse(read, 'fasta'):
        pfam_acc = record.id.split('/')[0]
        orgid = meta[meta['Entry name'] == pfam_acc]['Organism ID']
        taxid = orgid.tolist()
        if taxid == []:
            print(orgid)
        print(taxid)
       # organism = ncbi.get_taxid_translator([taxid])
```
