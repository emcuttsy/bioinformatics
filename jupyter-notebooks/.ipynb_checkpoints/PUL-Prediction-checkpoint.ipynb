```python
gff_dfs = {} # gene metadata from gff file
labels = ['sequence', 'source', 'feature', 'start', 'end', 'score', 'strand', 'phase', 'attributes']
for d in os.listdir():
    file = d + '/' + d + '.gff'
    df = pd.read_table(file, names=labels, skiprows=[0])
    splits = df['attributes'].str.split(';')
    ID = splits.str[0].str.split('=').str[1]
    locus_tag = splits.str[1].str.split('=').str[1]
    product = splits.str[2].str.split('=').str[1]
    df['id'] = ID
    df['locus tag'] = locus_tag
    df['product'] = product
   # df['product'] = product
    gff_dfs[int(d)] = df

gff_dfs[2806310567]
```

```python
seq_dfs = {}
for d in os.listdir():
    ids = []
    descs = []
    seqs = []
    file = d + '/' + d + '.genes.faa'
    with open(file, 'r') as fasta:
        for record in SeqIO.parse(fasta, 'fasta'):
            ids.append(record.id)
            descs.append(record.description)
            seqs.append(record.seq)
    df = pd.DataFrame({'id': ids, 'desc': descs, 'seq': seqs})
    seq_dfs[int(d)] = df
    print(df)
```

```python
pfam_dfs = {}
for d in os.listdir():
    file = d + '/' + d + '.pfam.tab.txt'
    df = pd.read_table(file)
    pfam_dfs[int(d)] = df
```

```python
genes_dfs = {}
for d in os.listdir():
    gff = gff_dfs[int(d)]
    seq = seq_dfs[int(d)]
    ids = seq['id']
    seqs = seq['seq']
    starts = []
    ends = []
    lens = []
    products = []
    strands = []
    locus_tags = []
    for index, row in seq.iterrows():
        entry = gff[gff['id'] == row['id']]
        starts.append(entry['start'].tolist()[0])
        ends.append(entry['end'].tolist()[0])
        lens.append(len(row['seq']))
        products.append(entry['product'].tolist()[0])
        strands.append(entry['strand'].tolist()[0])
        locus_tags.append(entry['locus tag'].tolist()[0])
    df = pd.DataFrame({'id': ids})
    df['locus tag'] = locus_tags
    df['start'] = starts
    df['end'] = ends
    df['length'] = lens
    df['strand'] = strands
    df['product'] = products
    df['seq'] = seqs
    genes_dfs[int(d)] = df
df.head()
```

```python
def get_gene_pfams(orgid, geneid):
    return pfam_dfs[orgid][pfam_dfs[orgid]['gene_oid'] == geneid]

def get_domain_end(pfamsdf, pfam):
    return max(pfamsdf[pfamsdf['pfam_id'] == pfam]['subj_end'])
```

```python
def isSusC(orgid, geneid):
    accepted_fams = ['pfam13715', 'pfam07715', 'pfam00593', 'pfam07660']
    pfams = get_gene_pfams(orgid, geneid)
    
    # if missing one of PF13715 or PF07715, it is not susC-like
    if 'pfam13715' not in pfams['pfam_id'].values or 'pfam07715' not in pfams['pfam_id'].values:
        return False
    
    # if contains non-allowed domains, it is not susC-like
    for fam in pfams['pfam_id']:
        if fam not in accepted_fams:
            return False

    # if PF13715 after PF07715 then it is not susC-like
    if get_domain_end(pfams, 'pfam07715') < get_domain_end(pfams, 'pfam13715'):
        return False
    
    # if PF07715 after PF00593 (if present), it is not susC-like
    elif not pfams[pfams['pfam_id'] == 'pfam00593'].empty:
        if get_domain_end(pfams, 'pfam00593') < get_domain_end(pfams, 'pfam07715'):
            return False
    
    return True

def isSusD(orgid, geneid):
    accepted_fams = ['pfam07980', 'pfam12741', 'pfam12771', 'pfam14322']
    pfams = get_gene_pfams(orgid, geneid)
    # if contains non-allowed domains, it is not susD-like
    for fam in pfams['pfam_id']:
        if fam not in accepted_fams:
            return False
    
    if len(pfams) > 2:
        return False
    else:
        return True 
```

```python
print(isSusC(2806310602, 2806874483))
print(get_gene_pfams(2806310602, 2806874483))
```

```python
susD = []
susC = []
for index, row in genes_dfs[2806310602].iterrows():
    geneid = int(row['id'])
    susD.append(isSusD(2806310602, geneid))
    susC.append(isSusC(2806310602, geneid))

genes_dfs[2806310602]['susD'] = susD
genes_dfs[2806310602]['susC'] = susC
genes_dfs[2806310602].head()
```
