```python
pfam_dfs = {}
for d in os.listdir():
    file = d + '/' + d + '.pfam.tab.txt'
    df = pd.read_table(file)
    pfam_dfs[int(d)] = df
```

```python
def get_gene_pfams(orgid, geneid):
    return pfam_dfs[orgid][pfam_dfs[orgid]['gene_oid'] == geneid]

def get_domain_end(pfamsdf, pfam):
    return max(pfamsdf[pfamsdf['pfam_id'] == pfam]['subj_end'])
```

```python
def isSusC(orgid, geneid):
    accepted_fams = ['pfam13715', 'pfam07715', 'pfam00593', 'pfam07660']
    pfams = get_gene_pfams(orgid, geneid)
    
    # if missing one of PF13715 or PF07715, it is not susC-like
    if 'pfam13715' not in pfams['pfam_id'].values or 'pfam07715' not in pfams['pfam_id'].values:
        return False
    
    # if contains non-allowed domains, it is not susC-like
    for fam in pfams['pfam_id']:
        if fam not in accepted_fams:
            return False

    # if PF13715 after PF07715 then it is not susC-like
    if get_domain_end(pfams, 'pfam07715') < get_domain_end(pfams, 'pfam13715'):
        return False
    
    # if PF07715 after PF00593 (if present), it is not susC-like
    elif not pfams[pfams['pfam_id'] == 'pfam00593'].empty:
        if get_domain_end(pfams, 'pfam00593') < get_domain_end(pfams, 'pfam07715'):
            return False
    
    return True

def isSusD(orgid, geneid):
    accepted_fams = ['pfam07980', 'pfam12741', 'pfam12771', 'pfam14322']
    pfams = get_gene_pfams(orgid, geneid)
    
    # if contains non-allowed domains, it is not susD-like
    for fam in pfams['pfam_id']:
        if fam not in accepted_fams:
            return False
    
    if len(pfams) > 2:
        return False
    
    return True 
```

```python
print(isSusC(2806310602, 2806874483))
print(get_gene_pfams(2806310602, 2806874483))
```

```python
count = 0
ls = []
for gene in pfam_dfs[2806310602]['gene_oid']:
    if isSusD(2806310602, gene):
        ls.append(gene)
        count +=1
print(len(set(ls)))
print(set(ls))
```
