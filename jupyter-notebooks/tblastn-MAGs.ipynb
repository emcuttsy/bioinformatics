```python
dfs = {}
header = ['query acc.ver', 'subject acc.ver', '% identity', 'alignment length', 'mismatches',\
          'gap opens', 'q. start', 'q. end', 's. start', 's. end', 'evalue', 'bit score']
for outfile in os.listdir():
    if outfile.endswith('.out'):
        name = outfile.split('.')[0].split('vs_')[1]
        dfs[name] = pd.read_table(outfile, names=header)
```

```python
os.listdir()
```

```python
# filter everything with an e value higher than 0.05
len_query = 1038
filtered_dfs = {}
for MAG in dfs.keys():
    df = dfs[MAG]
    filtered_df = df[df['evalue'] <= 0.05]
    filtered_df['coverage %'] =  (df['q. end'] - df['q. start'] + 1) / len_query
    filtered_df = filtered_df[filtered_df['coverage %'] >= .60]
    filtered_df['MAG'] = [MAG]*len(filtered_df)
    filtered_dfs[MAG] = filtered_df
```

```python
pd.concat(filtered_dfs.values())
```

```python
def write_translations(name, nucfasta, df):
    with open('/home/ecutts/Shark-Bay-MAGs/carrsulfatase_hits_' + name + '.faa', "w") as out:
        seqs = []
        for index, row in df.iterrows():
            start = int(row['s. start'])-1
            end = int(row['s. end'])-1
            seq = row['subject acc.ver']
            with open(nucfasta, "r") as handle:
                for record in SeqIO.parse(handle, "fasta"):
                    if record.id == seq:
                        nucseq = record.seq
                        if end < start:
                            start = len(nucseq) - start -1 
                            end = len(nucseq) - end -1
                            nucsub = nucseq.reverse_complement()[start:end]
                            prot = nucsub.translate()
                        else: 
                            nucsub = nucseq[start:end]
                            prot = nucsub.translate()
            neat_name = name.replace('_', ' ')
            rec = SeqRecord(prot, id=seq + '_trns',name='carr sulfatase hit',description='putative carrageenan sulfatase homolog ' + \
                        ' [' + neat_name + ']')
            seqs.append(rec)
        if len(seqs)>0:
            print(seqs)
            SeqIO.write(seqs, out, 'fasta')
```

```python
for MAG in filtered_dfs.keys():
    name = MAG
    df = filtered_dfs[MAG]
    nucfasta = '/home/ecutts/Shark-Bay-MAGs/nucleotide-fastas/SB_biofilm_' + name + '.fa'
    if len(df) >0:
        write_translations(name, nucfasta, df)
```

```python
counts = {}
rprots = pd.read_csv('/home/ecutts/metadata/rprots.txt', header = None)[0].tolist()
for faa in os.listdir():
    if faa.endswith('.faa'):
        name = faa.split('rprots_')[1].split('.faa')[0]
        counts[name] = dict.fromkeys(rprots)
        for prot in counts[name]:
            counts[name][prot] = 0
        with open(faa, 'r') as readfile:
            for record in SeqIO.parse(readfile, 'fasta'):
                prot = record.description.split('Protein ')[1].split(' [')[0]
                print(prot)
                counts[name][prot] += 1
                print(counts[name])
```

```python
for MAG in counts:
    print(MAG + ' count:' + str(sum(counts[MAG].values())))
    for rprot in counts[MAG].keys():
        if counts[MAG][rprot] != 1:
            print(rprot + ': ' + str(counts[MAG][rprot]))
    print('---')

    
```
