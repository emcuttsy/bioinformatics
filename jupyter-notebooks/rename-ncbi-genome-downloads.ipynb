```python
def read_report(report):
    assembly_report = open(report, 'r')

    while True:
        line = assembly_report.readline()
        if not line:
            break

        if 'Organism name' in line: 
            name = line.split('Organism name:  ')[1][:-1]
            name = name.split(' (')[0].replace(' ', '_')

        if 'GenBank assembly accession' in line:
            genbank_acc = line.split(' GenBank assembly accession: ')[1][:-1]

        if 'Taxid' in line:
            taxid = line.split('Taxid:          ')[1][:-1]
    
    return {'name': name, 'taxid': taxid, 'genbank': genbank_acc}

def find_report(d):
    for f in os.listdir(d):
        if f.endswith('assembly_report.txt'):
            return f
```

```python
for d in os.listdir():
    original = find_report(d).split('_assembly_report.txt')[0]
    info = read_report(d + '/' + find_report(d))
    rename = info['genbank'] + '_' + info['taxid']
    
    try:
        os.makedirs(destdir + '/' + rename)
    except:
        continue
        
    for f in os.listdir(d):
        try:
            extension = f.split(original)[1]
            filename = rename + extension
            print(filename)
        except:
            filename = f
        
        shutil.copyfile(d + '/' + f, destdir + '/' + rename + '/' + filename)
```
